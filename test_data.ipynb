```python
import pandas as pd
import random
```

```python
df = pd.read_csv('dataset/dataset.csv')
positive_data = df[df['is_duplicate'] == 1]
negative_data = df[df['is_duplicate'] == 0]
```

```python
triplets = []
for i in range(min(positive_data.shape[0],negative_data.shape[0])):
    anchor = positive_data.iloc[i]['question1_p']
    positive = positive_data.iloc[i]['question2_p']
    num = random.randint(1,2)
    negative = negative_data.iloc[i][f'question{num}_p']
    triplets.append((anchor,positive,negative))
```

```python
new_df = pd.DataFrame(triplets,columns=['anchor','positive','negative'])
new_df.head()
```

```python
new_df.to_csv('triplet.csv',index=False)
```
